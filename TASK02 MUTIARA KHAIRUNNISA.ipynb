```python
import csv 
import numpy as np
import matplotlib.pyplot as plt

# Mengatur font yang digunakan
plt.rcParams['font.size'] = 12

# Membuat list kosong untuk menyimpan isi file csv
time = []
velocity = []

# Impor dan baca data yang diberikan
filename = 'C:\\Tugas2_PrakMetNum_ForMutiara.csv'

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)

    # Baca setiap baris dalam csv
    for row in csvreader:
        time.append(float(row[0]))
        velocity.append(float(row[1]))

# Mendefinisikan interval waktu dan velocity
time = np.array(time)
velocity = np.array(velocity)

# Proses integrasi dengan aturan trapezoidal 
def integrate_fx(f, x, dx):
    # Buat array untuk menyimpan hasil integrasi
    hasil = np.zeros_like(x)

    # Loop untuk x untuk menghitung integrasi kumulatif
    for i in range(1, len(x)):
        trapz = dx * ((f[i-1] + f[i]) / 2)  
        hasil[i] = hasil[i-1] + trapz

    return hasil

# Proses diferensiasi dengan central differensiation
def central_difference(f, x, dx):
    # Buat array untuk menyimpan hasil integrasi
    hasil = np.zeros_like(f)

     # Loop untuk menghitung turunan menggunakan central difference
    for i in range(1, len(f)-1):
        hasil[i] = (f[i+1] - f[i-1]) / (2 * dx)
    return hasil

# Mendefinisikan dx
dx = 0.01

# Mendefinisikan rumus displacement (jarak)
displacement = integrate_fx(velocity, time, dx)

# Menghitung percepatan (acceleration) dari velocity terhadap time menggunakan central difference
acceleration = central_difference(velocity, time, dx)

# Plot Grafik Time vs Displacement
plt.plot(time, displacement, color='blue')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.title('Grafik Displacement')
plt.grid(True)
plt.show()

# Plot Grafik Time vs Velocity
plt.plot(time, velocity, color='green')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Grafik Velocity')
plt.grid(True)
plt.show()

# Plot Grafik Time vs Acceleration
plt.plot(time, acceleration, color='red') 
plt.xlabel('Time')
plt.ylabel('Acceleration')
plt.title('Grafik Acceleration')
plt.grid(True)
plt.show()
```
