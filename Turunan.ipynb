Turunan dalam numerik

Finite Diference 
1. Forward diff
df(x)/dx = (f(x+ deltax) - f(x)) / delta x

2. Backward diff
df(x)/dx = (f(x) - f(x-deltax)) / delta x

3. Central diff 
df(x)/dx = (f(x+ deltax) - f(x-delta x)) / 2 * delta x

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 20

def f(x):
    return x**3 + 2*x**2 + 3*x +4

def df_dx(x):
    return 3*x**2 + 4*x + 3

def frwd_diff(f, x, dx):
    return (f(x + dx) - f(x)) / dx

x = np.linspace(0, 5, 1000)

y = f(x)

dx = 0.1

y_df = df_dx(x)

y_frwd = frwd_diff(f, x, dx)

plt.plot(x, y, color='blue', label='f(x)')
plt.plot(x, y_df, color='green', label='df_dx')
plt.plot(x, y_frwd, color='red', label='forward')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 20

def f(x):
    return x**3 + 2*x**2 + 3*x +4

def df_dx(x):
    return 3*x**2 + 4*x + 3

def back_diff(f, x, dx):
    return (f(x) - f(x - dx)) / dx

x = np.linspace(0, 5, 1000)

y = f(x)

dx = 0.1

y_df = df_dx(x)

y_back = back_diff(f, x, dx)

plt.plot(x, y, color='blue', label='f(x)')
plt.plot(x, y_df, color='green', label='df_dx')
plt.plot(x, y_back, color='red', label='backdiff')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 16

def f(x):
    return x**3 + 2*x**2 + 3*x +4

def df_dx(x):
    return 3*x**2 + 4*x + 3

def frwd_diff(f, x, dx):
    return (f(x + dx) - f(x)) / dx

def ctrl_diff(f, x, dx):
    return (f(x + dx) - f(x - dx)) / (2*dx)

def back_diff(f, x, dx):
    return (f(x) - f(x - dx)) / dx

x = np.linspace(0, 5, 1000)

y = f(x)

dx = 0.8

y_df = df_dx(x)

y_frwd = frwd_diff(f, x, dx)

y_ctrl = ctrl_diff(f, x, dx)

y_back = back_diff(f, x, dx)

plt.plot(x, y, color='blue', label='f(x)')
plt.plot(x, y_df, color='black', label='df_dx')
plt.plot(x, y_frwd, color='red', label='forward diff')
plt.plot(x, y_ctrl, color='yellow', label='central diff', linestyle='--')
plt.plot(x, y_back, color='purple', label='backdiff')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

Aturan Trapezoidal untuk Integrasi

membagi kurva menjadi bentuk trapesium kecil kecil

```python
import numpy as np 

def f(x):
    return x**3 + 2*x**2 + 3*x +4

#interval integrasi 
a = 0
b = 2

# Jumlah nilai x antara a dan b (jumlah trapesium)
n = 10000

dx = (b - a) / n

x = np.linspace(a, b, n+1)

y = f(x)

integral = (dx/2) * (np.sum(y) - y[0] - y[-1]) #dari indeks ke 0 hingga indeks terakhir (dihitung dr belakang)

print(f"Integral numerik dari {a} ke {b}, adalah: {integral}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 10

def f(x):
    return x**3 + 2*x**2 + 3*x +4
def if_dx(x):
    return x**4/4 + 2*x**3/3 + 3*x**2/2 + 4*x
def integrate_fx(f, x, dx):  #mnghitung integral kumulatif
    #buat array untuk menyimpan hasil integrasi
    hasil = np.zeros_like(x)

    for i in range(1, len(x)):
        trapz = dx * ((f(x[i-1]) +f(x[i])) / 2) # menghitung aturan trapezoidal
        hasil[i] = hasil[i-1] + trapz

    return hasil

x = np.linspace(0, 5, 1000)

dx = x[1] - x[0]

y = f(x)

y_if = if_dx(x)

y_int = integrate_fx(f, x, dx)

plt.figure(figsize=(7, 5))
plt.plot(x, y, color='blue', label='fungsi f(x)')
plt.plot(x, y_if, color='black', label='integral f(x)dx')
plt.plot(x, y_int, color='red', label='trapezoidal rule', linestyle='--')
plt.xlabel('x')
plt.ylabel('Integral f(x)')
plt.legend()
plt.show()
```
